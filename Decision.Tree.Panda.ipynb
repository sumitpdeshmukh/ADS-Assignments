```python

import pandas as pd
from sklearn import tree
from sklearn.preprocessing import Imputer 
from IPython.display import Image

train_df = pd.read_csv("NewChurn70.csv")

train_df["CHURN"] = train_df["CHURN"].apply(lambda churn: 0 if churn == 'LEAVE' else 1)
train_df["COLLEGE"] = train_df["COLLEGE"].apply(lambda educated: 0 if educated == 'zero' else 1)

row_index = train_df.REPORTED_SATISFACTION == 'very_sat'
train_df.loc[row_index, "REPORTED_SATISFACTION"] = 1
row_index = train_df.REPORTED_SATISFACTION == 'sat'
train_df.loc[row_index, "REPORTED_SATISFACTION"] = 2
row_index = train_df.REPORTED_SATISFACTION == 'avg'
train_df.loc[row_index, "REPORTED_SATISFACTION"] = 3
row_index = train_df.REPORTED_SATISFACTION == 'unsat'
train_df.loc[row_index, "REPORTED_SATISFACTION"] = 4
row_index = train_df.REPORTED_SATISFACTION == 'very_unsat'
train_df.loc[row_index, "REPORTED_SATISFACTION"] = 5

row_index = train_df.REPORTED_USAGE_LEVEL == 'very_high'
train_df.loc[row_index, "REPORTED_USAGE_LEVEL"] = 1
row_index = train_df.REPORTED_USAGE_LEVEL == 'high'
train_df.loc[row_index, "REPORTED_USAGE_LEVEL"] = 2
row_index = train_df.REPORTED_USAGE_LEVEL == 'avg'
train_df.loc[row_index, "REPORTED_USAGE_LEVEL"] = 3
row_index = train_df.REPORTED_USAGE_LEVEL == 'little'
train_df.loc[row_index, "REPORTED_USAGE_LEVEL"] = 4
row_index = train_df.REPORTED_USAGE_LEVEL == 'very_little'
train_df.loc[row_index, "REPORTED_USAGE_LEVEL"] = 5

row_index = train_df.CONSIDERING_CHANGE_OF_PLAN == 'considering'
train_df.loc[row_index, "CONSIDERING_CHANGE_OF_PLAN"] = 1
row_index = train_df.CONSIDERING_CHANGE_OF_PLAN == 'actively_looking_into_it'
train_df.loc[row_index, "CONSIDERING_CHANGE_OF_PLAN"] = 2
row_index = train_df.CONSIDERING_CHANGE_OF_PLAN == 'no'
train_df.loc[row_index, "CONSIDERING_CHANGE_OF_PLAN"] = 3
row_index = train_df.CONSIDERING_CHANGE_OF_PLAN == 'never_thought'
train_df.loc[row_index, "CONSIDERING_CHANGE_OF_PLAN"] = 4
row_index = train_df.CONSIDERING_CHANGE_OF_PLAN == 'perhaps'
train_df.loc[row_index, "CONSIDERING_CHANGE_OF_PLAN"] = 5

y = targets = labels = train_df["CHURN"].values

columns = ["COLLEGE", "INCOME", "OVERAGE", "LEFTOVER", "HOUSE", "HANDSET_PRICE",
           "OVER_15MINS_CALLS_PER_MONTH", "AVERAGE_CALL_DURATION", "REPORTED_SATISFACTION", 
           "REPORTED_USAGE_LEVEL","CONSIDERING_CHANGE_OF_PLAN"]
features = train_df[list(columns)].values



imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
X = imp.fit_transform(features)

clf = tree.DecisionTreeClassifier(criterion="entropy", max_depth=4)
clf = clf.fit(X, y)

from sklearn.externals.six import StringIO
with open("tree.dot", 'w') as f:
    f = tree.export_graphviz(clf, out_file=f, feature_names=columns)

    

test_df = pd.read_csv("NewChurn70.csv")

test_df["CHURN"] = test_df["CHURN"].apply(lambda churn: 0 if churn == 'LEAVE' else 1)
test_df["COLLEGE"] = test_df["COLLEGE"].apply(lambda educated: 0 if educated == 'zero' else 1)

row_index = test_df.REPORTED_SATISFACTION == 'very_sat'
test_df.loc[row_index, "REPORTED_SATISFACTION"] = 1
row_index = test_df.REPORTED_SATISFACTION == 'sat'
test_df.loc[row_index, "REPORTED_SATISFACTION"] = 2
row_index = test_df.REPORTED_SATISFACTION == 'avg'
test_df.loc[row_index, "REPORTED_SATISFACTION"] = 3
row_index = test_df.REPORTED_SATISFACTION == 'unsat'
test_df.loc[row_index, "REPORTED_SATISFACTION"] = 4
row_index = test_df.REPORTED_SATISFACTION == 'very_unsat'
test_df.loc[row_index, "REPORTED_SATISFACTION"] = 5

row_index = test_df.REPORTED_USAGE_LEVEL == 'very_high'
test_df.loc[row_index, "REPORTED_USAGE_LEVEL"] = 1
row_index = test_df.REPORTED_USAGE_LEVEL == 'high'
test_df.loc[row_index, "REPORTED_USAGE_LEVEL"] = 2
row_index = test_df.REPORTED_USAGE_LEVEL == 'avg'
test_df.loc[row_index, "REPORTED_USAGE_LEVEL"] = 3
row_index = test_df.REPORTED_USAGE_LEVEL == 'little'
test_df.loc[row_index, "REPORTED_USAGE_LEVEL"] = 4
row_index = test_df.REPORTED_USAGE_LEVEL == 'very_little'
test_df.loc[row_index, "REPORTED_USAGE_LEVEL"] = 5

row_index = test_df.CONSIDERING_CHANGE_OF_PLAN == 'considering'
test_df.loc[row_index, "CONSIDERING_CHANGE_OF_PLAN"] = 1
row_index = test_df.CONSIDERING_CHANGE_OF_PLAN == 'actively_looking_into_it'
test_df.loc[row_index, "CONSIDERING_CHANGE_OF_PLAN"] = 2
row_index = test_df.CONSIDERING_CHANGE_OF_PLAN == 'no'
test_df.loc[row_index, "CONSIDERING_CHANGE_OF_PLAN"] = 3
row_index = test_df.CONSIDERING_CHANGE_OF_PLAN == 'never_thought'
test_df.loc[row_index, "CONSIDERING_CHANGE_OF_PLAN"] = 4
row_index = test_df.CONSIDERING_CHANGE_OF_PLAN == 'perhaps'
test_df.loc[row_index, "CONSIDERING_CHANGE_OF_PLAN"] = 5    

features = test_df[list(columns)].values
imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
X = imp.fit_transform(features)
predictlist = clf.predict(X)
actualList = test_df["CHURN"]
correct_guess = 0
wrong_guess = 0
index = 0
for pr in predictlist:
    if(pr == actualList[index]):
        correct_guess += 1
    else:
        wrong_guess += 1
    index +=1

print("Accuracy in % ",(correct_guess/index)*100)
print("Error in % ",(wrong_guess/index)*100)
     

```
