```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
input = open('Churn.csv')
count = -1
index = 0
college = []
binary_college = []
income = []
overage = []
leftover = []
house = []
handset = []
overuse = []
avuse = []
satisfaction = []
value_satisfaction = []
uselvl = []
value_uselvl = []
consideration = []
value_consideration = []
churn = []
binary_churn = []
result = []
coeff_val = []
for line in input:
    if count == -1:
        count = 0
    else:    
        newline = line.split(",")
        college.append(newline[0])
        if newline[0] == "zero":
            binary_college.append(0)
        else:
            binary_college.append(1)
        income.append(int(newline[1]))
        overage.append(int(newline[2]))
        leftover.append(int(newline[3]))
        house.append(int(newline[4]))
        handset.append(int(newline[5]))
        overuse.append(int(newline[6]))
        avuse.append(int(newline[7]))
        satisfaction.append(newline[8])
        
        if newline[8] == "very_sat": 
            value_satisfaction.append(5)
        elif newline[8] == "sat":
            value_satisfaction.append(4)
        elif newline[8] == "avg":
            value_satisfaction.append(3)
        elif newline[8] == "unsat":
            value_satisfaction.append(2)
        elif newline[8] == "very_unsat":
            value_satisfaction.append(1)
        
        uselvl.append(newline[9])
        if newline[9] == "very_little": 
            value_uselvl.append(5)
        elif newline[9] == "little":
            value_uselvl.append(4)
        elif newline[9] == "avg":
            value_uselvl.append(3)
        elif newline[9] == "high":
            value_uselvl.append(2)
        elif newline[9] == "very_high":
            value_uselvl.append(1)
            
        consideration.append(newline[10])
        if newline[10] == "actively_looking_into_it": 
            value_consideration.append(5)
        elif newline[10] == "considering":
            value_consideration.append(4)
        elif newline[10] == "no":
            value_consideration.append(3)
        elif newline[10] == "never_thought":
            value_consideration.append(2)
        elif newline[10] == "perhaps":
            value_consideration.append(1)
        
        newline = newline[11].split("\n")
        churn.append(newline[0])
        if newline[0] == "LEAVE":
            binary_churn.append(0)
        else:
            binary_churn.append(1)
        ++count
#############################################################################


coeff_val = stats.pearsonr(binary_college,binary_churn)
print("Correlation Coefficient between\n")
print("COLLEGE and CHURN ", coeff_val[0])
result.append(coeff_val[0])
coeff_val = stats.pearsonr(income,binary_churn)
print("INCOME and CHURN ", coeff_val[0])
result.append(coeff_val[0])
coeff_val = stats.pearsonr(overage,binary_churn)
print("OVERAGE and CHURN ", coeff_val[0])
result.append(coeff_val[0])
coeff_val = stats.pearsonr(house,binary_churn)
print("HOUSE and CHURN ", coeff_val[0])
result.append(coeff_val[0])
coeff_val = stats.pearsonr(leftover,binary_churn)
print("LEFTOVER and CHURN ", coeff_val[0])
result.append(coeff_val[0])
coeff_val = stats.pearsonr(handset,binary_churn)
print("HANDSET and CHURN ", coeff_val[0])
result.append(coeff_val[0])
coeff_val = stats.pearsonr(overuse,binary_churn)
print("LONG CALLS and CHURN ", coeff_val[0])
result.append(coeff_val[0])
coeff_val = stats.pearsonr(avuse,binary_churn)
print("AVERAGE USE and CHURN ", coeff_val[0])
result.append(coeff_val[0])
coeff_val = stats.pearsonr(value_satisfaction,binary_churn)
print("SATISFACTION and CHURN ", coeff_val[0])
result.append(coeff_val[0])
coeff_val = stats.pearsonr(value_uselvl,binary_churn)
print("USAGE LEVEL and CHURN ", coeff_val[0])
result.append(coeff_val[0])
coeff_val = stats.pearsonr(value_consideration,binary_churn)
print("CONSIDERATION and CHURN ", coeff_val[0])
result.append(coeff_val[0])

result2 = []

for c in result:
    result2.append(abs(c))

width = 0.35
N = 11
fig = plt.figure()
ind = np.arange(N)
barlist = plt.bar(ind, result2, width, color='b')
cnt = 0

for val in result: 
    if val < 0:
        barlist[cnt].set_color('r')
    cnt = cnt + 1

plt.title("Correlation Coefficients (Independent Variable vs Churn)")
plt.ylabel('Absolute Correlation Coefficient')
plt.xlabel('Independent Variables (Features)')

plt.xticks(ind + width/2., ('College', 'Income','Average Overcharges', 'House', 'Average Leftover Minutes','Handset Price', 'Long Calls','Average Duration', 'Satisfaction', 'Usage Level', 'Consideration'))
plt.yticks(np.arange(0, 0.3, 0.05))
fig.autofmt_xdate()
plt.show()
```
